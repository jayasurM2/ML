```python

import pandas as pd
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/chico.csv"
df=pd.read_csv(url)
print(df.head(5))
print(df.shape)
fig=plt.figure(figsize=(8,6),dpi=800)
test1 = fig.add_subplot(2,2,1)#nrows, ncols, index
test2 = fig.add_subplot(2,2,4)
test3= fig.add_subplot(2,2,2)
#pos is a three-digit integer where the first, second, and third integer are nrows,ncols, index.
test1.boxplot(df["grade_test1"],showfliers=False)#method1   #showfliers=False(Hide outliers)
test2.boxplot(df["grade_test2"],showfliers=False)
test3.bar(df["grade_test2"],df["id"])
plot = df[['grade_test1', 'grade_test2']].plot(kind='box', title='boxplot', showmeans=True)#method2
plt.show()
```

```python
import numpy as np
x=np.linspace(0,10,100)
plt.subplot(2,2,1)
plt.plot(x-np.sin(x),"-")
plt.subplot(2,2,2)
plt.plot(x-np.cos(x),"*")
plt.subplot(2,2,3)
plt.plot(x-np.sin(x),"--")
plt.subplot(2,2,4)
plt.plot(x-np.cos(x),":")
plt.show()
```

```python
plt.plot(x,np.sin(x),color="green",linestyle="solid")
plt.plot(x,x+np.sin(x),color="blue",linestyle="dashed")
plt.plot(x,x-np.sin(x),color="#00ffff",linestyle="dotted")
plt.plot(x,x*np.sin(x),":k")
plt.show()
```
