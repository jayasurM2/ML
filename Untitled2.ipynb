```python
import numpy as np
import statistics as s
d1=[1,2,3,4,5,6,7,8]
print(np.average(d1))
print(np.var(d1))
print(np.std(d1))
print(np.mean(d1))
print(np.median(d1))
print(np.min(d1))
print(np.max(d1))
print(s.mode(d1))
print(np.max(d1)-np.min(d1))


#IQR using Numpy
Q1 = np.percentile(d1, 25, interpolation = 'midpoint') 
# Third quartile (Q3)
Q3 = np.percentile(d1, 75, interpolation = 'midpoint')
# Interquaritle range (IQR)
IQR = Q3 - Q1
print(IQR)
```
